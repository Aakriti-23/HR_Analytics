```python
df['DOB'] = pd.to_datetime(df['DOB'], format="%d-%m-%Y")
from datetime import datetime
current_year = datetime.now().year
df["Age"] = current_year - df["DOB"].dt.year
# df["Age"]

df["Age_Group"] = pd.cut(
    df["Age"],
    bins=[0, 25, 45, 60, 100],
    labels=["Young", "Adult", "Middle Age", "Senior"]
)


print(df["Age_Group"].value_counts())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))

# Average Age
df.groupby('Age_Group')['Age'].mean().plot(
    kind='bar',
    ax=axes[0],
    color=["#D8F3DC", "#B7E4C7", "#95D5B2", "#74C69D"]
)
axes[0].set_title('Average Age by Age Group')
axes[0].set_xlabel('Age Group')
axes[0].set_ylabel('Average Age')
axes[0].tick_params(axis='x', rotation=0)

# Performance Score
df.groupby('Age_Group')['Performance_Score_num'].mean().plot(
    kind='bar',
    ax=axes[1],
    color=["#D8F3DC", "#B7E4C7", "#95D5B2", "#74C69D"]
)
axes[1].set_title('Performance Score by Age Group')
axes[1].set_xlabel('Age Group')
axes[1].set_ylabel('Avg Performance Score')
axes[1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
```
